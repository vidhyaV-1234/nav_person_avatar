```python
# --- Configuration ---
STEPS = 1040
BATCH_SIZE = 6
FP_16 = True
PRETRAINED_MODEL = "runwayml/stable-diffusion-v1-5"
INSTANCE_DIR = "/content/data_example"
OUTPUT_DIR = "/content/output"
PROMPT = "ktn"
LEARNING_RATE = 1e-4
TRAIN_TEXT_ENCODER = True
LEARNING_RATE_TEXT_ENCODER = 5e-5

# Derived config
NEW_LEARNING_RATE = LEARNING_RATE / BATCH_SIZE
NEW_LEARNING_RATE_TEXT_ENCODER = LEARNING_RATE_TEXT_ENCODER / BATCH_SIZE
fp_16_arg = "fp16" if FP_16 else "no"

# Clean up old checkpoints
!rm -rf {INSTANCE_DIR}/.ipynb_checkpoints

# --- Training command ---
base_cmd = (
    f"accelerate launch /content/diffusers/examples/dreambooth/train_dreambooth_lora.py "
    f"--pretrained_model_name_or_path=\"{PRETRAINED_MODEL}\" "
    f"--instance_data_dir=\"{INSTANCE_DIR}\" "
    f"--output_dir=\"{OUTPUT_DIR}\" "
    f"--instance_prompt=\"{PROMPT}\" "
    f"--resolution=512 "
    f"--use_8bit_adam "
    f"--mixed_precision=\"{fp_16_arg}\" "
    f"--train_batch_size=1 "
    f"--gradient_accumulation_steps=1 "
    f"--learning_rate={NEW_LEARNING_RATE} "
    f"--lr_scheduler=\"constant\" "
    f"--lr_warmup_steps=0 "
    f"--max_train_steps={STEPS} "
    f"--lora_rank=16 "
    f"--learning_rate_text={NEW_LEARNING_RATE_TEXT_ENCODER} "
)

if TRAIN_TEXT_ENCODER:
    base_cmd += "--train_text_encoder "

# Run training
!{base_cmd}
```

```python
import os
from safetensors.torch import load_file, save_file
from diffusers import StableDiffusionPipeline
import torch

# Step 1: Split pytorch_lora_weights.safetensors
lora_weights_path = "/content/output/pytorch_lora_weights.safetensors"
output_dir = "/content/output"
unet_weights_path = os.path.join(output_dir, "lora_weights.safetensors")
text_encoder_weights_path = os.path.join(output_dir, "lora_weights.text_encoder.safetensors")

if not os.path.exists(lora_weights_path):
    raise FileNotFoundError(f"LoRA weights file not found at {lora_weights_path}")

os.makedirs(output_dir, exist_ok=True)
weights = load_file(lora_weights_path)



# Filter weights for UNet and Text Encoder
unet_weights = {k: v for k, v in weights.items() if any(term in k.lower() for term in ["unet", "down_blocks", "up_blocks", "mid_block", "attn1", "attn2", "cross_attention", "attention"])}
text_encoder_weights = {k: v for k, v in weights.items() if any(term in k.lower() for term in ["text_encoder", "clip", "transformer", "self_attn"])}


# Save weights if they exist
if unet_weights:
    save_file(unet_weights, unet_weights_path)
else:
    print("No UNet weights found")

if text_encoder_weights:
    save_file(text_encoder_weights, text_encoder_weights_path)
else:
    print("No Text Encoder weights found")

# Step 2: Load pipeline and apply LoRA weights
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16
).to("cuda")

# Apply LoRA weights using diffusers' native support
if os.path.exists(unet_weights_path):
    pipe.load_lora_weights(output_dir, weight_name="lora_weights.safetensors")
else:
    print("Skipping UNet LoRA weights application (file not found)")

if os.path.exists(text_encoder_weights_path):
    pipe.load_lora_weights(output_dir, weight_name="lora_weights.text_encoder.safetensors")
else:
    print("Skipping Text Encoder LoRA weights application (file not found)")


INFERENCE_PROMPT = "a photo of ktn "
GUIDANCE = 7.5

try:
    image = pipe(INFERENCE_PROMPT, num_inference_steps=50, guidance_scale=GUIDANCE).images[0]
    image.save("/content/output/generated_image.png")
    print("Image generated and saved to /content/output/generated_image.png")
    image.show()
except Exception as e:
    print(f"Inference failed: {e}")
```
